## Bike count data

<div class="alert alert-block alert-info">

1. Load all csv files into data frames. For each, create a variable and name the variable according to the respective filename. Familiarize yourself with the data.

</div>

```python
import pandas as pd

# Define file paths
data_path = "C:/Users/Admin/Downloads/DS400 Data Science Project Management WS23-24/Assignments/Assignment2/data/"

# Load datasets
bike_counts = pd.read_csv(f"{data_path}bike_counts.csv")
counter_sites = pd.read_csv(f"{data_path}counter_sites.csv")
holidays = pd.read_csv(f"{data_path}holidays.csv")
weather = pd.read_csv(f"{data_path}weather.csv")

# Display the first few rows of each dataset
print(bike_counts.head())
print(counter_sites.head())
print(holidays.head())
print(weather.head())
```

```python
# Function to extract column names and data types
def get_column_info(df):
    data_info = pd.DataFrame({
        "Column_Name": df.columns,
        "Data_Type": df.dtypes.astype(str)
    })
    return data_info

# Extract column names and data types for each dataset
bike_counts_info = get_column_info(bike_counts)
counter_sites_info = get_column_info(counter_sites)
holidays_info = get_column_info(holidays)
weather_info = get_column_info(weather)

# Display column names and data types
print("Bike Counts Info")
print(bike_counts_info)

print("Counter Sites Info")
print(counter_sites_info)

print("Holidays Info")
print(holidays_info)

print("Weather Info")
print(weather_info)
```

2. In python, convert all date or datetime columns to a date/datetime datatype. We have not covered this during the lecture, but you can easily use your favorite search engine to learn more about it. In R, this should have been done automatically when importing the csv.

```python
# Convert time and date columns to appropriate data types
bike_counts['time'] = pd.to_datetime(bike_counts['time'], format='%Y-%m-%d %H:%M:%S')
holidays['date'] = pd.to_datetime(holidays['date'], format='%Y-%m-%d')
weather['time'] = pd.to_datetime(weather['time'], format='%Y-%m-%d %H:%M:%S')

# Display the data types to verify changes
print(bike_counts.dtypes)
print(holidays.dtypes)
print(weather.dtypes)
```

<div class="alert alert-block alert-info">

3. Aggregate the `bike_counts` dataset by `counter_site` such that the number of bikes per hour is now the sum of the `bike_counts` across all channels. Call the resulting object `bike_counts_agg`.

</div>

```python
# Aggregate bike counts by counter_site_id and time
bike_counts_agg = bike_counts.groupby(['counter_site_id', 'time'], as_index=False).agg({'bike_count': 'sum'})

# Display the aggregated dataset
print(bike_counts_agg.head())
```

## Missing values

<div class="alert alert-block alert-info">

5. For the columns contained in `bike_counts` and `counter_sites` there should be no missing values. For the `holidays` columns, however, you will encounter a lot of missing values after the join. Briefly describe the reason for this. Can you come up with a reasonable approach to fill these missing values? Implement your approach.

</div>

```python
# Function to check for missing values and return column names with missing values
def check_missing_values(df):
    missing_cols = df.columns[df.isnull().any()].tolist()
    return missing_cols

# Check for missing values in each table
bike_counts_missing = check_missing_values(bike_counts_agg)
counter_sites_missing = check_missing_values(counter_sites)
holidays_missing = check_missing_values(holidays)
weather_missing = check_missing_values(weather)
merged_missing = check_missing_values(merged)

# Display columns with missing values
print("Bike Counts columns with missing values:")
print(bike_counts_missing)

print("Counter Sites columns with missing values:")
print(counter_sites_missing)

print("Holidays columns with missing values:")
print(holidays_missing)

print("Weather columns with missing values:")
print(weather_missing)

print("Merged columns with missing values:")
print(merged_missing)

# Fill missing values in holidays columns with 0
merged['semester_break'].fillna(0, inplace=True)
merged['school_holiday'].fillna(0, inplace=True)
merged['public_holiday'].fillna(0, inplace=True)

# Check for remaining missing values
merged_missing_after_fill = check_missing_values(merged)
print("Merged missing columns after fill:")
print(merged_missing_after_fill)
```

Hint: We haven't covered this in class, thus for the grouping part you might want to have a look at the API documentation of [{pandas}](https://pandas.pydata.org/docs/reference/api/pandas.Grouper.html) (python) or [{lubridate}](https://lubridate.tidyverse.org/reference/round_date.html) (R).

```python
# Aggregate bike counts to weekly counts
bike_counts['week'] = bike_counts['time'].dt.to_period('W').apply(lambda r: r.start_time)
bike_counts_weekly = bike_counts.groupby(['counter_site_id', 'week']).agg({'bike_count': 'sum'}).reset_index()

# Display the aggregated dataset
print(bike_counts_weekly.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Merge with counter_sites to get the counter site names
bike_counts_weekly = pd.merge(bike_counts_weekly, counter_sites, on='counter_site_id', how='left')

# Create faceted line plot
g = sns.FacetGrid(bike_counts_weekly, col="counter_site", col_wrap=1, height=4, aspect=2, sharex=True, sharey=True)
g.map(sns.lineplot, 'week', 'bike_count')

# Set axis labels and titles
g.set_axis_labels("Week", "Number of Bikes")
g.set_titles("{col_name}")
g.fig.suptitle("Weekly Bike Counts by Counter Site", y=1.02)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### B: Bike counts per hour and weekday

Next, we want to look for timetable-ish patterns.

<div class="alert alert-block alert-info">

10. Aggregate the `bike_counts` dataset to get the average number of bikes per hour separated by weekday (across all counter sites).

</div>

```python
# Extract weekday from the time column
bike_counts['weekday'] = bike_counts['time'].dt.day_name()

# Aggregate to get the average number of bikes per hour by weekday
avg_bikes_per_weekday = bike_counts.groupby('weekday')['bike_count'].mean().reset_index()

# Ensure the days of the week are ordered correctly
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
avg_bikes_per_weekday['weekday'] = pd.Categorical(avg_bikes_per_weekday['weekday'], categories=days_order, ordered=True)
avg_bikes_per_weekday = avg_bikes_per_weekday.sort_values('weekday')

# Display the aggregated dataset
print(avg_bikes_per_weekday)
```

<div class="alert alert-block alert-info">

11. Plot the resulting average bike count in a heatmap such that it looks like a timetable (`weekday` on the x-axis and `hour` on the y-axis).

</div>

```python
# Extract weekday and hour from the time column
bike_counts['weekday'] = bike_counts['time'].dt.day_name()
bike_counts['hour'] = bike_counts['time'].dt.hour

# Aggregate to get the average number of bikes per hour by weekday and hour
avg_bikes_per_weekday_hour = bike_counts.groupby(['weekday', 'hour'])['bike_count'].mean().reset_index()

# Ensure the days of the week are ordered correctly
days_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
avg_bikes_per_weekday_hour['weekday'] = pd.Categorical(avg_bikes_per_weekday_hour['weekday'], categories=days_order, ordered=True)
avg_bikes_per_weekday_hour = avg_bikes_per_weekday_hour.sort_values(['weekday', 'hour'])

# Pivot the dataframe for heatmap
heatmap_data = avg_bikes_per_weekday_hour.pivot("hour", "weekday", "bike_count")

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="Reds", annot=True, fmt=".1f", cbar_kws={'label': 'Avg Bike Count'})
plt.title("Average Bike Count by Weekday and Hour")
plt.xlabel("Weekday")
plt.ylabel("Hour")
plt.show()
```

## Fixed effects regression

Finally, we want to regress the number of bikes per hour (`bike_count`) on the other variables.

<div class="alert alert-block alert-info">

13. Which variables should be included as fixed effects? What about the `time` column? Since we are not estimating a time series model, it does not make sense to include it just as it is. However, there is certainly relevant information contained in the timestamp. By referencing to the results of the last section, state which information of the timestamp could be relevant/predictive. Extract this information by creating new variables in the data frame.    
    *(Hint: You can use `merged.time.dt.month` (python) or `month(time)` from the {lubridate} package (R) to extract for example the month of a timestamp)*    
    Would you include these variables as numeric variables or fixed effects?

</div>



```python
# Extract relevant information from the timestamp
merged['hour'] = merged['time'].dt.hour
merged['day_of_week'] = merged['time'].dt.day_name()
merged['month'] = merged['time'].dt.month_name()

# View the new variables
print(merged.head())

import statsmodels.api as sm
import statsmodels.formula.api as smf

# Fit the regression model
model = smf.ols('bike_count ~ hour + day_of_week + month + temperature + windspeed + humidity + heaven + rain + snow + thunderstorms + fog + semester_break + school_holiday + public_holiday + counter_site_id', data=merged).fit()

# Summary of the model
print(model.summary())
```

<div class="alert alert-block alert-info">

14. Make a train-test-split. In this case, just choose a single week and counter site of your choice as test set and use the rest for the estimation. Make sure your code is reproducible. Estimate the model on the training sample. To make sure the model outputs only positive values, we apply a trick: Instead of using the `bike_count` in levels as dependent variable, use the square root. You may specify this in the formula directly.

</div>

```python
import numpy as np
np.random.seed(123)
# Select a specific week and counter site for the test set
test_start_date = "2021-07-01"
test_end_date = "2021-07-07"
test_counter_site_id = 100003358

test_set = merged[(merged['time'].dt.date >= pd.to_datetime(test_start_date)) &
                  (merged['time'].dt.date <= pd.to_datetime(test_end_date)) &
                  (merged['counter_site_id'] == test_counter_site_id)]

# Create the training set
train_set = merged[~((merged['time'].dt.date >= pd.to_datetime(test_start_date)) &
                     (merged['time'].dt.date <= pd.to_datetime(test_end_date)) &
                     (merged['counter_site_id'] == test_counter_site_id))]

# Fit the regression model on the training set using the square root of bike_count as the dependent variable
model = smf.ols('np.sqrt(bike_count) ~ hour + day_of_week + month + temperature + windspeed + humidity + heaven + rain + snow + thunderstorms + fog + semester_break + school_holiday + public_holiday + counter_site_id', data=train_set).fit()
```

<div class="alert alert-block alert-info">

15. Print the model summary and briefly summarize your findings.

</div>

```python
# Summary of the model
print(model.summary())
```

<div class="alert alert-block alert-info">

16. Make predictions on the test set. Remember, that we fitted the model to the square root of the `bike_counts`. Compute the mean absolute error of your predictions.

</div>

```python
# Make predictions on the test set
test_set['predicted_bike_count'] = np.square(model.predict(test_set))

# Compute the mean absolute error
mae = np.mean(np.abs(test_set['predicted_bike_count'] - test_set['bike_count']))
print("Mean Absolute Error:", mae)
```

<div class="alert alert-block alert-info">

17. Plot the prediction on your test set as a time series. Also add the 95% confidence interval of your prediction as filled area as well as the actual values as points. The resulting figure should look like the one in the instructions PDF. Note that is also includes the mean squared error in the legend entry for the prediction. Summarize your findings.

</div>

```python
# Calculate confidence intervals
pred_ci = model.get_prediction(test_set).conf_int()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(test_set['time'], test_set['predicted_bike_count'], color='blue', label=f'Prediction (MAE={round(mae, 2)})', linestyle='solid')
plt.scatter(test_set['time'], test_set['bike_count'], color='red', label='Actual', s=10)
plt.fill_between(test_set['time'], pred_ci[:, 0], pred_ci[:, 1], color='lightblue', alpha=0.5, label='95% CI')
plt.suptitle('Prediction of bike counts per hour', fontsize=16)
plt.title('Counter Site: Fuß- & Radtunnel Südportal - Derendinger Allee')
plt.xlabel('Date')
plt.ylabel('Bike Count')
plt.legend()
plt.grid(False)
plt.show()
```
