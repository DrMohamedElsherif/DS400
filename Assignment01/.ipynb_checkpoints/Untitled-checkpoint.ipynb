```python
import os
from pathlib import Path

# Set the current working directory to the directory where this script or notebook is located
os.chdir(Path().resolve())

# Verify working directory
print("Current Working Directory:", os.getcwd())


import numpy as np
import matplotlib.pyplot as plt

# Define the receptive field function components
def excitatory(x):
    return 2 * np.exp(-0.5 * x**2)

def inhibitory(x):
    return np.exp(-0.5 * (x**2 / 2**2))

# Define the total receptive field function
def receptive_field(x):
    return excitatory(x) - inhibitory(x)

# Define the x range for plotting
x = np.linspace(-5, 5, 500)

# Calculate the function values
exc = excitatory(x)
inhib = inhibitory(x)
rf = receptive_field(x)

# Plot the Gaussians and their sum
plt.figure(figsize=(10, 6))
plt.plot(x, exc, label='Excitatory', color='blue')
plt.plot(x, inhib, label='Inhibitory', color='red')
plt.plot(x, rf, label='Receptive Field (Sum)', color='green', linestyle='dashed')

# Add labels and title
plt.title('Excitatory and Inhibitory Mechanisms and their Sum')
plt.xlabel('x')
plt.ylabel('Receptive Field Value')
plt.legend()
plt.grid(False)
# Save the plot as an image
plt.savefig("receptive_field_plot.png", dpi=300)  # Save as PNG with 300 dpi for high resolution
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from scipy.integrate import quad

# Define the receptive field function as a difference of Gaussians
def receptive_field(r, sigma_center=1, sigma_surround=2, k=1.5):
    center = np.exp(-r**2 / (2 * sigma_center**2))
    surround = k * np.exp(-r**2 / (2 * sigma_surround**2))
    return center - surround

# Generate the 2D plot for the receptive field with circular stimulus
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)

# Receptive field values
Z = receptive_field(R)

# Plot the receptive field overlaid with a circular spot of light
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot the receptive field
im = ax[0].imshow(Z, extent=(-5, 5, -5, 5), origin='lower', cmap='bwr', norm=Normalize(vmin=-1, vmax=1))
ax[0].set_title("Receptive Field with Stimulus")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

# Overlay a circular stimulus with radius s
s = 0.25  # Radius of the stimulus
circle = plt.Circle((0, 0), s, color= 'yellow', alpha=0.3, label=f"Stimulus Radius $s = {s}$")
ax[0].add_artist(circle)
ax[0].legend()

# Calculate and plot the tuning curve
def response(s):
    integrand = lambda r: 2 * np.pi * r * receptive_field(r)  # Integrate in polar coordinates
    return quad(integrand, 0, s)[0]

s_values = np.linspace(0, 5, 100)
R_values = np.array([response(s) for s in s_values])

ax[1].plot(s_values, R_values, color='blue')
ax[1].set_title("Tuning Curve")
ax[1].set_xlabel("Stimulus Radius $s$")
ax[1].set_ylabel("Total Response $R(s)$")
ax[1].axhline(0, color='black', linestyle='--', linewidth=0.5)  # Zero response line

plt.colorbar(im, ax=ax[0], orientation="vertical", label="Receptive Field Strength")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the receptive field function with adjusted parameters
def receptive_field(r, sigma_center=1, sigma_surround=2.5, k=0.8):
    center = np.exp(-r**2 / (2 * sigma_center**2))
    surround = k * np.exp(-r**2 / (2 * sigma_surround**2))
    return center - surround

# Calculate the response R(s) by integrating the receptive field within a radius s
def response(s, sigma_center=1, sigma_surround=2.5, k=0.19):
    integrand = lambda r: 2 * np.pi * r * receptive_field(r, sigma_center, sigma_surround, k)
    return quad(integrand, 0, s)[0]

# Generate a range of stimulus sizes
s_values = np.linspace(0, 5, 100)  # stimulus radius values from 0 to 5
R_values = np.array([response(s) for s in s_values])  # compute the response for each radius

# Plot the tuning curve
plt.figure(figsize=(8, 6))
plt.plot(s_values, R_values, color='blue', linewidth=2)
plt.title("Bell-Shaped Tuning Curve")
plt.xlabel("Stimulus Radius $s$")
plt.ylabel("Total Response $R(s)$")
plt.axhline(0, color='black', linestyle='--', linewidth=0.5)  # Zero response line
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the receptive field function for 2D
def receptive_field(r, sigma_center=1, sigma_surround=2.5, k=0.19):
    # Excitatory center (Gaussian)
    center = np.exp(-r**2 / (2 * sigma_center**2))
    # Inhibitory surround (Gaussian with scaling factor k)
    surround = k * np.exp(-r**2 / (2 * sigma_surround**2))
    return center - surround

# Generate the 2D grid of x and y coordinates
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Calculate the radial distance R from the center for each point
R = np.sqrt(X**2 + Y**2)

# Calculate the receptive field values Z at each point
Z = receptive_field(R)

# Plotting the 2D receptive field
plt.figure(figsize=(8, 6))

# Plot the receptive field using a contour plot
contour = plt.contourf(X, Y, Z, levels=50, cmap='RdBu_r')
plt.colorbar(contour)  # Add color bar to show the scale of receptive field values

# Highlight a circular stimulus with a specific radius
stimulus_radius = 0.5
circle = plt.Circle((0, 0), stimulus_radius, color='black', fill=False, linestyle='--', linewidth=2)
plt.gca().add_artist(circle)

# Titles and labels
plt.title("Receptive Field with Circular Stimulus")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show grid and set equal aspect ratio
plt.grid(False)
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the receptive field function for 2D
def receptive_field(r, sigma_center=0.5, sigma_surround=2.5, k=0.19):
    # Excitatory center (Gaussian)
    center = np.exp(-r**2 / (2 * sigma_center**2))
    # Inhibitory surround (Gaussian with scaling factor k)
    surround = k * np.exp(-r**2 / (2 * sigma_surround**2))
    return center - surround

# Calculate the response R(s) by integrating the receptive field within a radius s
def response(s, sigma_center=0.5, sigma_surround=2.5, k=0.19):
    integrand = lambda r: 2 * np.pi * r * receptive_field(r, sigma_center, sigma_surround, k)
    return quad(integrand, 0, s)[0]

# Generate the 2D grid of x and y coordinates
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Calculate the radial distance R from the center for each point
R = np.sqrt(X**2 + Y**2)

# Calculate the receptive field values Z at each point
Z = receptive_field(R)

# Define stimulus radii for tuning curve (stimulus size range)
stimulus_radii = np.linspace(0, 5, 50)

# Calculate the total response for each stimulus radius
R_values = np.array([response(s) for s in stimulus_radii])

# Plotting the 2D receptive field
fig, ax1 = plt.subplots(figsize=(20, 16))

# Plot the receptive field using a contour plot
contour = ax1.contourf(X, Y, Z, levels=50, cmap='RdBu_r')
fig.colorbar(contour)  # Add color bar to show the scale of receptive field values

# Highlight a circular stimulus with a specific radius (initial stimulus radius)
stimulus_radius = 1
circle = plt.Circle((0, 0), stimulus_radius, color='black', fill=False, linestyle='--', linewidth=2)
ax1.add_artist(circle)

# Titles and labels for the receptive field plot
ax1.set_title("Receptive Field with Circular Stimulus")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.grid(False)
ax1.set_aspect('equal', adjustable='box')

# Plotting the tuning curve
ax2 = ax1.twinx()  # Create a second y-axis for the tuning curve
ax2.plot(stimulus_radii, R_values, color='blue', linewidth=2, label="Tuning Curve")
ax2.set_ylabel("Response $R(s)$")
ax2.set_xlabel("Stimulus Radius $s$")
ax2.set_ylim([0, np.max(R_values)*1.1])  # Adjust y-axis to fit the response values

# Plot an initial stimulus size (optional)
ax2.axvline(x=stimulus_radius, color='red', linestyle='--', label=f"Stimulus Radius = {stimulus_radius}")

# Add a legend for the tuning curve
ax2.legend(loc='upper right')

# Show the plot
plt.show()
```
