<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
##### I- listings_muc.csv
    
**listings_muc.csv is arguably NOT tidy for the following reasons:**

- <u>**Main Reason**</u> The 'host_id' and 'host_name' columns form a different type of observational unit and are categorical variables that won't change, Therefore we can create a separate table to store this information. This task wasn't explicitly mentioned in the assignment, so I didn't include it as part of my solution below.
- The enitre 'neighbourhood_group' column and major portion of 'license' column have missing values, which questions the purpose of having both variables included for analysis. To rectify this, we should remove both. 
- Missing values in columns 'last_review' and its current values are stored as an object datatype, requiring conversion to datetime for analysis. However since dates of 'last_review' offers limited value in current analysis unless feature engineered (e.g., into "days since last review"), I opted to drop it from the dataset.
- Missing values in column 'reviews_per_month' columns when 'number_of_reviews' has value of 0. To rectify this, we can replace those missing values with 0. 
- The column 'id' needs to be converted from float in to integer or object (to avoid issues with decimals). In R, no need to perform such step as R will automatically remove the '.0'.
- The columns 'name' and 'host_name' (which has 1 missing value) seem irrelevant to analysis and can be removed.</div>

```python
# Drop 'name', 'host_name', 'neighbourhood_group', 'last_review', and 'license' columns
AirBNB_listings_muc.drop(columns=['name', 'host_name', 'neighbourhood_group', 'last_review', 'license'], inplace=True)

# Convert the 'id' column to integer 
AirBNB_listings_muc['id'] = AirBNB_listings_muc['id'].astype('int64')

# Set 'reviews_per_month' to 0 where 'number_of_reviews' is 0
AirBNB_listings_muc.loc[AirBNB_listings_muc['number_of_reviews'] == 0, 'reviews_per_month'] = 0
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
##### II- calendar_muc.csv
    
**calendar_muc.csv is NOT tidy because:**

- <u>**Main Reason**</u> The 'min_max_nights' information are two different variables placed in one column, and each cell has two values. To rectify this, we should separate into two columns: 'minimum_nights' and 'maximum_nights'.
- The columns 'price' and 'date' need to be converted to float/numeric (after removing the '$' sign) and datetime format respectively.
- The column 'listing_id' need to be converted to integer or object (to avoid issues with decimals). In R, no need to perform such step as R will automatically remove the '.0'.
- The enitre 'adjusted_price' column has missing values, which questions the purpose of having this variable included for analysis. To rectify this, we should remove it.
</div>

```python
# Splitting 'min_max_nights' into 'min_nights' and 'max_nights'
AirBNB_calendar_muc[['minimum_nights', 'maximum_nights']] = AirBNB_calendar_muc['min_max_nights'].str.split('-', expand=True)

# Converting new columns to integers
AirBNB_calendar_muc['minimum_nights'] = AirBNB_calendar_muc['minimum_nights'].astype('int64')
AirBNB_calendar_muc['maximum_nights'] = AirBNB_calendar_muc['maximum_nights'].astype('int64')

# Dropping the original 'min_max_nights' column
AirBNB_calendar_muc = AirBNB_calendar_muc.drop(columns=['min_max_nights'])

# Convert 'price' to numeric by removing the dollar sign and converting to float
AirBNB_calendar_muc['price'] = AirBNB_calendar_muc['price'].replace('[\$,]', '', regex=True).astype(float)

# Convert 'date' to datetime format
AirBNB_calendar_muc['date'] = pd.to_datetime(AirBNB_calendar_muc['date'], format='%Y-%m-%d')

# Convert 'listing_id' to integer 
AirBNB_calendar_muc['listing_id'] = AirBNB_calendar_muc['listing_id'].astype('int64')

# Drop the 'adjusted_price' column as it has no values
AirBNB_calendar_muc = AirBNB_calendar_muc.drop(columns=['adjusted_price'])
```

### Merging the data

<div class="alert alert-block alert-info">
7. In the calendar data, change the name of the listing_id column to id.</div>

```python
# Rename 'listing_id' to 'id' in the AirBNB_calendar_muc DataFrame
AirBNB_calendar_muc.rename(columns={'listing_id': 'id'}, inplace=True)
```

<div class="alert alert-block alert-info">
8.  Aggregate the calendar data to the listing level. The variable available should be transformed into the number of available days per listing. The variable date should be discarded. The variables price, minimum_nights, and maximum_nights should be aggregated to the listing level..</div>

```python
# Aggregate the calendar data
calendar_agg = AirBNB_calendar_muc.groupby('id').agg(
    available=('available', 'sum'),  # Count total available days
    price=('price', 'last'),               # Get the last price for each listing
    minimum_nights=('minimum_nights', 'min'),  # Minimum nights required
    maximum_nights=('maximum_nights', 'max')   # Maximum nights allowed
).reset_index()
```

<div class="alert alert-block alert-info">
9. Merge the aggregated calendar and the listings data by id. Call the resulting data frame merged and
print its shape.</div>

```python
# Merge with the listings dataset
merged = pd.merge(AirBNB_listings_muc, calendar_agg, on='id', how='left')

print(merged.shape)
merged.head(5)
```

```python
# Check for missing values in each column of the merged DataFrame
missing_values = merged.isna().sum()

# Print the missing values for each column
print("Missing values in each column:")
print(missing_values)
```

<div class="alert alert-block alert-info">
11. Regress the number of available days on the price and the minimum_nights. Interpret the results. What would you say – is this a good model?</div>

```python
import statsmodels.api as sm

# Define the independent variables (price and minimum_nights) and add a constant for the intercept
X = merged[['price', 'minimum_nights']]
X = sm.add_constant(X)       

# Define the dependent variable (available)
y = merged['available']

# Fit the OLS regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
 #### **Interpretation:**

- <u>**Price Coefficient (0.0039)**</u> indicates that for each additional unit (i.e. $) increase in price, the number of available days is expected to increase by approximately 0.0039 days, holding the minimum nights constant. The coefficient is statistically significant (p = 0.001), suggesting that price has a positive effect on availability. However, the effect size is very small.
- <u>**Minimum Nights Coefficient (0.2371)**</u> indicates that for each additional night added to the minimum stay requirement, the number of available days is expected to increase by about 0.2371 days, holding price constant. This coefficient is also statistically significant (p < 0.001), suggesting that there is a positive relationship between minimum nights and availability.

**No the model is not good,** given the very low R-squared value (0.005), as the model explains very little of the variance in the number of available days. 
</div>

<div class="alert alert-block alert-info">
12. In a second model, regress the number of available days on the price and the minimum_nights, but include neighborhood and room type as fixed effects. Interpret the results. Do you see any problems with this model?</div>

```python
import statsmodels.formula.api as smf

# Define the formula
formula = 'available ~ price + minimum_nights + C(neighbourhood) + C(room_type)'

# Fit the regression model
second_model = smf.ols(formula=formula, data=merged).fit()
print(second_model.summary())
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
#### **Interpretation:**

- <u>**Price Coefficient (0.0044)**</u> indicates that for each additional unit (i.e. $) increase in price, the number of available days is expected to increase by approximately 0.0044 days, holding the minimum nights constant. The coefficient is statistically significant (p < 0.001), suggesting that price has a positive effect on availability. However, the effect size is still very small.
- <u>**Minimum Nights Coefficient (0.2667)**</u> indicates that for each additional night added to the minimum stay requirement, the number of available days is expected to increase by about 0.2667 days, holding price constant. This coefficient is also statistically significant (p < 0.001), suggesting that there is a positive relationship between minimum nights and availability.

- <u>**Neighborhood Effects**</u> shows several neighborhoods having negative coefficients such as Altstadt-Lehel (-37.06 days), Au-Haidhausen (-62.24 days), Bogenhausen (-40.27 day) and Hadern (-60.18 days) indicating fewer available days compared to the baseline neighborhood. These negative coefficients suggest that listings in these neighborhoods tend to have significantly fewer available days compared to the reference neighborhood. Neighborhoods such as Untergiesing-Harlaching and Tudering-Riem have coefficients close to zero, indicating minimal effect on availability.

- <u>**Room Type Effects**</u> shows coefficient 118.43 for Hotel room, suggesting that hotel rooms have significantly more available days than the baseline room type compared to other room types.

**The problems I see in this model** is that the R-squared value (0.034) is still low indicating that the model does not explain much of the variance in the dependent variable and that there may be other important variables not included in this model. The low R-squared value could also be due to presence of multicollinearity or other factors that violate assumptions of OLS which are: normally distributed residuals, homoscedasticity (constant variance of residuals), and independence of observations.
