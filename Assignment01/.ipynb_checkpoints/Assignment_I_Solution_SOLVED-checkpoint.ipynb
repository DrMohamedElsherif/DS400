<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
##### I- policies.csv
    
**policies.csv is NOT tidy for the following reasons:**

- The dataset features a column for each day of the calendar year. To streamline the data, we can consolidate these columns into a single one representing the date variable.

- The district type variable is currently split into two separate columns. To streamline this, we can merge both columns into a unified representation.

- Missing values in the dataset, particularly in the district type and stringency index columns.

Since 'district' and 'district_type' are categorical variables that won't change, we can create a separate table to store this information. This task wasn't explicitly mentioned in the assignment, so I didn't include it as part of my solution below."
</div>

```python
# Step 1: Split the 'district' column into 'district' and 'district_type'
policies[['district', 'district_type']] = policies['district'].str.split(',', expand=True).rename(columns={0: 'district', 1:'district_type'})

# Step 2: Iterate over the district types and create boolean columns to filter and update
for district_type in ['Stadt', 'Kreis']:
    boolean_col = policies[f'district_type={district_type}']
    policies.loc[boolean_col, 'district_type'] = district_type
    policies.drop(columns=[f'district_type={district_type}'], inplace=True)
    
# Step 3: Reshape the DataFrame so that date becomes a column instead of row
policies = policies.melt(id_vars=['district', 'district_type'], var_name='date', value_name='stringency_index')

# Step 4: Convert the 'date' column to datetime
policies['date'] = pd.to_datetime(policies['date'])

# Step 5 Drop rows where SI is missing
policies = policies.dropna(subset=['stringency_index'])

# Step 6: Drop rows with invalid dates
policies = policies.dropna(subset=['date'])

print(policies['district_type'].unique())
policies.head()
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
##### II- cases.csv
    
**cases.csv is considered NOT tidy because:**

- The date information is spread across two separate columns. To rectify this, we should combine these two columns into a single date column.
- There are missing values within the district_type variable in the dataset.
</div>

```python
# Step 1: Combine year_month and day into a single date column
cases['date'] = pd.to_datetime(cases.year_month.astype(str) + cases.day.astype(str).str.zfill(2))

# Step 2: Drop the separate 'year_month' and 'day' columns
cases.drop(columns=['year_month', 'day'], inplace=True)

print(cases['district_type'].unique())
cases.head()
```

### Merging the data

<div class="alert alert-block alert-info">

6. Merge the two tables by district, district_type and date. Only keep rows with a match in both
tables. In R, if you use {dplyr} make sure to fill missing values in the district_type column before
the merge. Otherwise it won’t work properly. Call the resulting data frame merged and print its shape.

</div>


```python
# Merge the two DataFrames by district, district_type, and date
merged = pd.merge(policies, cases, on=['district', 'district_type', 'date'], how='inner')

print("Shape of the merged DataFrame:", merged.shape)
```
