4. Create a barplot showing the number of athletes per nationality (from highest to lowest). What is wrong with this plot?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the number of athletes per nationality
nationality_counts = Athletes['nationality'].value_counts()

# Plot the barplot
plt.figure(figsize=(20, 6))
sns.barplot(
    x=nationality_counts.index,
    y=nationality_counts.values,
    order=nationality_counts.index[:], 
)
plt.xticks(rotation=90)
plt.xlabel('Nationality')
plt.ylabel('Number of Athletes')
plt.title('Number of Athletes per Nationality')
plt.tight_layout()
plt.show()
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    The barplot has several issues: a large number of categories can clutter the x-axis, especially if NaN or empty nationalities are not filtered out, and the dominance of a few nationalities may overshadow less-represented ones, making them nearly invisible. Additionally, displaying all categories can lead to a top-heavy visualization that lacks meaningful comparison.
</div>

<div class="alert alert-block alert-info">   
5. Make another barplot where only the most common nationalities are displayed (defined as having at least 100 athletes at the Olympics). The remaining athletes should be grouped into a “other” bar. Include the actual number of athletes in each bar.
</div>

```python
# Filter nationalities with at least 100 athletes
common_nationalities = nationality_counts[nationality_counts >= 100]
other = nationality_counts[nationality_counts < 100].sum()

# Create a new series with "Other" grouped
filtered_nationalities = common_nationalities.append(pd.Series({'Other': other}))

# Create the bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(filtered_nationalities.index, filtered_nationalities.values)

# Add counts above the bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), 
             ha='center', va='bottom', fontsize=10)

plt.title("Number of Athletes per Nationality (Grouped by ≥100 Athletes)", fontsize=14)
plt.xlabel("Nationality", fontsize=12)
plt.ylabel("Number of Athletes", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

<div class="alert alert-block alert-info">   
6. Using the medals dataset, select the top 5 countries with the most medals. Create a plot showing the number of medals on the y-axis and the countries on the x-axis. Color the bars per gender. To make the plot easier to read, include only athletes that identify themselves as male or female. Use a colorblind-friendly palette
</div>

```python
# Filter medals dataset to include only male and female athletes
medals_filtered = Medals[Medals['gender'].isin(['M', 'W'])]

# Calculate the top 5 countries with the most medals
top_countries = (
    medals_filtered.groupby('country_code')
    .size()
    .nlargest(5)
    .index
)

# Filter dataset for the top 5 countries
top_medals = medals_filtered[medals_filtered['country_code'].isin(top_countries)]

# Create the plot
plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_medals,
    x='country_code',
    y='medal_code',
    hue='gender',
    palette="colorblind",
    estimator=len,
    ci=None
)
plt.title("Number of Medals by Gender for Top 5 Countries")
plt.xlabel("Country")
plt.ylabel("Number of Medals")
plt.legend(title="Gender")
plt.show()
```

11. Regress the number of available days on the price and the minimum_nights. Interpret the results. What would you say – is this a good model?</div>

```python
import statsmodels.api as sm

# Define the independent variables (price and minimum_nights) and add a constant for the intercept
X = merged[['price', 'minimum_nights']]
X = sm.add_constant(X)       

# Define the dependent variable (available)
y = merged['available']

# Fit the OLS regression model
model = sm.OLS(y, X).fit()
print(model.summary())
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
 #### **Interpretation:**

- <u>**Price Coefficient (0.0039)**</u> indicates that for each additional unit (i.e. $) increase in price, the number of available days is expected to increase by approximately 0.0039 days, holding the minimum nights constant. The coefficient is statistically significant (p = 0.001), suggesting that price has a positive effect on availability. However, the effect size is very small.
- <u>**Minimum Nights Coefficient (0.2371)**</u> indicates that for each additional night added to the minimum stay requirement, the number of available days is expected to increase by about 0.2371 days, holding price constant. This coefficient is also statistically significant (p < 0.001), suggesting that there is a positive relationship between minimum nights and availability.

**No the model is not good,** given the very low R-squared value (0.005), as the model explains very little of the variance in the number of available days. 
</div>

12. In a second model, regress the number of available days on the price and the minimum_nights, but include neighborhood and room type as fixed effects. Interpret the results. Do you see any problems with this model?</div>

```python
import statsmodels.formula.api as smf

# Define the formula
formula = 'available ~ price + minimum_nights + C(neighbourhood) + C(room_type)'

# Fit the regression model
second_model = smf.ols(formula=formula, data=merged).fit()
print(second_model.summary())
```

<div class="alert alert-block alert-info" style="background-color: yellow;">   
    
#### **Interpretation:**

- <u>**Price Coefficient (0.0044)**</u> indicates that for each additional unit (i.e. $) increase in price, the number of available days is expected to increase by approximately 0.0044 days, holding the minimum nights constant. The coefficient is statistically significant (p < 0.001), suggesting that price has a positive effect on availability. However, the effect size is still very small.
- <u>**Minimum Nights Coefficient (0.2667)**</u> indicates that for each additional night added to the minimum stay requirement, the number of available days is expected to increase by about 0.2667 days, holding price constant. This coefficient is also statistically significant (p < 0.001), suggesting that there is a positive relationship between minimum nights and availability.

- <u>**Neighborhood Effects**</u> shows several neighborhoods having negative coefficients such as Altstadt-Lehel (-37.06 days), Au-Haidhausen (-62.24 days), Bogenhausen (-40.27 day) and Hadern (-60.18 days) indicating fewer available days compared to the baseline neighborhood. These negative coefficients suggest that listings in these neighborhoods tend to have significantly fewer available days compared to the reference neighborhood. Neighborhoods such as Untergiesing-Harlaching and Tudering-Riem have coefficients close to zero, indicating minimal effect on availability.

- <u>**Room Type Effects**</u> shows coefficient 118.43 for Hotel room, suggesting that hotel rooms have significantly more available days than the baseline room type compared to other room types.

**The problems I see in this model** is that the R-squared value (0.034) is still low indicating that the model does not explain much of the variance in the dependent variable and that there may be other important variables not included in this model. The low R-squared value could also be due to presence of multicollinearity or other factors that violate assumptions of OLS which are: normally distributed residuals, homoscedasticity (constant variance of residuals), and independence of observations.
